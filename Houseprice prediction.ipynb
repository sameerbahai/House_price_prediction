```python
price_deciding_col=['bedrooms','bathrooms','sqft_living','floors','waterfront','view','grade','sqft_above', 'sqft_basement','lat','sqft_living15']
```

```python
X=df[price_deciding_col].values
Y=df['price'].values
```

```python
from sklearn.preprocessing import StandardScaler,MinMaxScaler
ind_mn=MinMaxScaler()
X=ind_mn.fit_transform(X)
mn=MinMaxScaler()
y=mn.fit_transform(Y.reshape(-1,1))
X.shape,y.shape

```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(X_train,Y_train)
pred=reg.predict(X_test)
y_pred=reg.predict(X_train)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
MSE=mean_squared_error(Y_test,pred)
print("mean squared error:",MSE)
RMSE=np.sqrt(MSE)
print("root mean squared error:",RMSE)
score=r2_score(Y_test,pred)
print(" r2_score :",score)
```

```python
fig, ax = plt.subplots()
ax.scatter(Y_train, y_pred, label="Training Set")
ax.scatter(Y_test,pred, label="testing Set")
ax.plot([0, max(Y_train)], [0, max(Y_train)], 'r--')
ax.set_xlabel("Actual house price")
ax.set_ylabel("Predicted house price")
ax.set_title("linearRegression")
ax.legend()
plt.show()
```
